```python
import pandas as pd

print("library Import Done cuyy")
```

```python
f = 'dinkes-od_17448_jml_penderita_diabetes_melitus_brdsrkn_kabupatenko_v2_data.csv'
df = pd.read_csv(f)
print ("Load Data sukses bree ")
```

```python
# 14.Kelompokkan data berdasarkan tahun dan hitung total jumlah_penderita_dm 
# per tahun di Jawa Barat.

print("\n--- Total Penderita DM per Tahun ---")
total_per_tahun = df.groupby('tahun')['jumlah_penderita_dm'].sum()
print(total_per_tahun)
```

```python
# 15. Kelompokkan data berdasarkan nama_kabupaten_kota dan hitung 
# rata-rata jumlah_penderita_dm tiap kabupaten/kota (jika terdapat 
# lebih dari satu tahun).

print("\n--- Rata-rata Penderita DM per Kabupaten/Kota ---")
avg_per_kota = df.groupby('nama_kabupaten_kota')['jumlah_penderita_dm'].mean()
print(avg_per_kota)
```

```python
# 16. Tentukan kabupaten/kota dengan total jumlah_penderita_dm tertinggi 
# dan terendah di seluruh tahun.

print("\n--- Kabupaten/Kota dengan Total DM Tertinggi & Terendah ---")
total_per_kota = df.groupby('nama_kabupaten_kota')['jumlah_penderita_dm'].sum()

print("Tertinggi :", total_per_kota.idxmax(), total_per_kota.max())
print("Terendah  :", total_per_kota.idxmin(), total_per_kota.min())
```

```python
# 17. Buat kolom baru bernama kategori_dm dengan aturan berbasis jumlah_penderita_dm:
# < 50.000 → "Rendah"
# 50.000–99.999 → "Sedang"
# = 100.000 → "Tinggi"

print("\n--- Membuat Kolom Kategori DM ---")

def tentukan_kategori(jumlah):
    if jumlah >= 100000:
        return "Tinggi"
    elif jumlah >= 50000:
        return "Sedang"
    else:
        return "Rendah"

df['kategori_dm'] = df['jumlah_penderita_dm'].apply(tentukan_kategori)
print(df[['nama_kabupaten_kota', 'jumlah_penderita_dm', 'kategori_dm']].head())
```

```python
# 18. Untuk setiap tahun, hitung persentase jumlah penderita DM per kabupaten/kota 
# terhadap total jumlah penderita DM di tahun tersebut, lalu simpan dalam kolom baru 
# bernama persentase_tahun.Persentase per tahun

print("\n--- Persentase Kontribusi per Tahun ---")

df['persentase_tahun'] = (
    df['jumlah_penderita_dm'] /
    df.groupby('tahun')['jumlah_penderita_dm'].transform('sum')
) * 100

print(df[['tahun', 'nama_kabupaten_kota',
          'jumlah_penderita_dm', 'persentase_tahun']].head())
```

```python
# 19. Tabel Ringkas (Tahun, Total Penderita, Jumlah Kota Lapor)
summary_table = df.groupby('tahun').agg({
    'jumlah_penderita_dm': 'sum',
    'nama_kabupaten_kota': 'count'
}).reset_index()

summary_table.columns = ['Tahun', 'Total_Penderita', 'Jml_Kota_Lapor']
print("\n--- Tabel Ringkasan ---")
print(summary_table)
```
