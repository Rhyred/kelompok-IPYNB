```python
import pandas as pd

# Load data ulang (biar pasti)
nama_file = 'disperkim-od_18378_jml_sampah_terangkut_ke_tempat_pembuangan_akhir_tpa__k_data.csv'
df = pd.read_csv(nama_file)
```

```python
print("=== JAWABAN SOAL 3.1 ===")
df_2021 = df[df['tahun'] == 2021].copy()
print("Berhasil filter data tahun 2021.")
print(df_2021.head()) # Cek 5 data pertama
```

```python
print("=== JAWABAN SOAL 3.2 ===")
top_5_2021 = df_2021.sort_values(by='jumlah_sampah', ascending=False).head(5)
print("5 Data Teratas (sudah diurutkan):")
print(top_5_2021)
```

```python
print("=== JAWABAN SOAL 3.3 ===")
hasil_soal_3 = top_5_2021[['nama_kabupaten_kota', 'jumlah_sampah', 'tahun']]
print("5 Wilayah dengan Sampah Terbanyak Tahun 2021:")
print(hasil_soal_3)
```

```python
print("=== JAWABAN SOAL 3.4 ===")
juara_1 = hasil_soal_3.iloc[0]['nama_kabupaten_kota']
jumlah_juara_1 = hasil_soal_3.iloc[0]['jumlah_sampah']
print(f"\nKesimpulan: Wilayah dengan sampah terbanyak adalah {juara_1} ({jumlah_juara_1} Ton/Hari).")
print("-" * 30)
```

```python

print("=== SOAL 4: Statistik Sampah per Kota ===")

nama_file = 'disperkim-od_18378_jml_sampah_terangkut_ke_tempat_pembuangan_akhir_tpa__k_data.csv'
df = pd.read_csv(nama_file)

statistik_kota = df.groupby('nama_kabupaten_kota')['jumlah_sampah'].agg(['mean', 'median'])

df_statistik = statistik_kota.reset_index()
df_statistik.columns = ['nama_kabupaten_kota', 'mean_jumlah_sampah', 'median_jumlah_sampah']

df_statistik_urut = df_statistik.sort_values('mean_jumlah_sampah', ascending=False)
print("3 Kabupaten/Kota dengan Rata-rata Sampah Terbesar:")
print(df_statistik_urut.head(3))

print("\nCatatan: Nilai rata-rata Purwakarta sangat tinggi karena ada satu tahun datanya melonjak drastis (outlier).")
```

```python
print("=== JAWABAN SOAL 5 ===")
cek_2015 = df[df['tahun'] == 2015]

if cek_2015.empty:
    print("Info: Data tahun 2015 TIDAK DITEMUKAN. Menggunakan tahun 2016 sebagai pembanding awal.")
    tahun_awal = 2016
else:
    tahun_awal = 2015

tahun_akhir = 2023
```

```python
print("=== JAWABAN SOAL 5.1 ===")
total_awal = df[df['tahun'] == tahun_awal]['jumlah_sampah'].sum()
total_akhir = df[df['tahun'] == tahun_akhir]['jumlah_sampah'].sum()
print("tahun awal: ", total_awal)
print("tahun akhir: ", total_akhir)
```

```python
print("=== JAWABAN SOAL 5.2 ===")
selisih = total_akhir - total_awal
persentase = (selisih / total_awal) * 100

print(f"Total Sampah {tahun_awal}: {total_awal:,.2f} Ton/Hari")
print(f"Total Sampah {tahun_akhir}: {total_akhir:,.2f} Ton/Hari")
print(f"Selisih: {selisih:,.2f} Ton/Hari")
print(f"Persentase Perubahan: {persentase:.2f}%")
```

```python
print("=== JAWABAN SOAL 5.3 ===")
if selisih > 0:
    print("Kesimpulan: Terjadi KENAIKAN jumlah sampah.")
else:
    print("Kesimpulan: Terjadi PENURUNAN jumlah sampah yang terangkut.")
```

```python
df_2021 = df[df['tahun'] == 2021].copy()
print("Data tahun 2021 berhasil diambil.")
```

```python

total_jabar_2021 = df_2021['jumlah_sampah'].sum()
print(f"Total Sampah Jabar 2021: {total_jabar_2021:,.2f} Ton/Hari")
```

```python

df_2021['persentase_kontribusi'] = (df_2021['jumlah_sampah'] / total_jabar_2021) * 100
print("Kolom persentase berhasil ditambahkan.")
```

```python

top_10 = df_2021.sort_values('persentase_kontribusi', ascending=False).head(10)
hasil_akhir = top_10[['nama_kabupaten_kota', 'jumlah_sampah', 'persentase_kontribusi']]

print("\n10 Kabupaten/Kota dengan Kontribusi Terbesar:")
print(hasil_akhir)
```

```python

top_3_total = hasil_akhir.head(3)['persentase_kontribusi'].sum()
print("\n=== KESIMPULAN ===")
print(f"Tiga wilayah teratas (Kota Bandung, Depok, Bekasi) menyumbang sekitar {top_3_total:.2f}% dari total sampah.")
print("Analisis: Kontribusi mereka SANGAT SIGNIFIKAN karena hanya dari 3 kota ini saja sudah mencakup hampir sepertiga total sampah se-Jawa Barat.")    
```

```python

pivot_sampah = df.pivot_table(index='nama_kabupaten_kota', 
                              columns='tahun', 
                              values='jumlah_sampah')

print("Pivot Table berhasil dibuat!")
```

```python

jumlah_nan = pivot_sampah.isna().sum().sum()
print(f"Jumlah data kosong (NaN) ditemukan: {jumlah_nan}")

kota_contoh = 'KABUPATEN SUMEDANG'
data_awal = pivot_sampah.loc[kota_contoh]
print(f"\nData awal {kota_contoh} (Tahun 2018 harusnya NaN/Kosong):")
print(data_awal)
```

```python
pivot_sampah_clean = pivot_sampah.fillna(0)
```

```python
print(f"\nData {kota_contoh} SETELAH diisi 0:")
print(pivot_sampah_clean.loc[kota_contoh])
```

```python
print("\n=== ANALISIS ===")
if data_awal.isna().any():
    tahun_hilang = data_awal[data_awal.isna()].index.tolist()
    print(f"Ya, terdapat tahun yang tidak memiliki data untuk {kota_contoh}, yaitu tahun {tahun_hilang}.")
    print("Setelah proses cleaning, nilai NaN tersebut telah berubah menjadi 0.0.")
else:
    print(f"Tidak ada data yang hilang untuk {kota_contoh}.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
nama_file = 'disperkim-od_18378_jml_sampah_terangkut_ke_tempat_pembuangan_akhir_tpa__k_data.csv'
df = pd.read_csv(nama_file)
```

```python
kota_pilihan = 'KOTA BANDUNG'
df_kota = df[df['nama_kabupaten_kota'] == kota_pilihan].sort_values('tahun')
```

```python
plt.plot(df_kota['tahun'], 
         df_kota['jumlah_sampah'], 
         marker='o', 
         linestyle='-', 
         color='blue', 
         linewidth=2,
         label='Jumlah Sampah')
plt.title(f'Tren Jumlah Sampah Terangkut di {kota_pilihan} (2016-2023)', fontsize=14)
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Jumlah Sampah (Ton/Hari)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(df_kota['tahun']) # Memastikan semua tahun muncul
plt.legend()

hasil_gambar = 'grafik_tren_sampah_bandung.png'
plt.savefig(hasil_gambar, dpi=300) # dpi=300 agar gambar jernih (HD)
plt.show()
print(f"Berhasil! Grafik sudah disimpan dengan nama file: {hasil_gambar}")
```
