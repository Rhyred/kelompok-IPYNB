```python

# 14. Total jumlah penderita DM per tahun di Jawa Barat
print("\n--- 14. Total Penderita per Tahun ---")
total_per_tahun = df.groupby('tahun')['jumlah_penderita_dm'].sum()
print(total_per_tahun)

# 15. Rata-rata penderita DM tiap kabupaten/kota (lintas tahun)
print("\n--- 15. Rata-rata Penderita per Kota/Kab ---")
avg_per_kota = df.groupby('nama_kabupaten_kota')['jumlah_penderita_dm'].mean()
print(avg_per_kota)

# 16. Kabupaten/Kota dengan Total Tertinggi dan Terendah (Seluruh Tahun)
print("\n--- 16. Kota dengan Total DM Tertinggi & Terendah ---")
# Hitung dulu total per kota
total_per_kota = df.groupby('nama_kabupaten_kota')['jumlah_penderita_dm'].sum()

# Mencari nama kota dengan nilai max dan min
kota_tertinggi = total_per_kota.idxmax() # Mengambil index (nama kota) max
nilai_tertinggi = total_per_kota.max()
kota_terendah = total_per_kota.idxmin()  # Mengambil index (nama kota) min
nilai_terendah = total_per_kota.min()

print(f"Tertinggi: {kota_tertinggi} ({nilai_tertinggi} orang)")
print(f"Terendah : {kota_terendah} ({nilai_terendah} orang)")

# 17. Buat kolom kategori_dm (Rendah, Sedang, Tinggi)
print("\n--- 17. Membuat Kolom Kategori ---")

def tentukan_kategori(jumlah):
    if jumlah >= 100000:
        return 'Tinggi'
    elif jumlah >= 50000:
        return 'Sedang'
    else:
        return 'Rendah'

# Terapkan fungsi ke kolom jumlah_penderita_dm
df['kategori_dm'] = df['jumlah_penderita_dm'].apply(tentukan_kategori)

# Cek hasil (tampilkan 5 data acak)
print(df[['nama_kabupaten_kota', 'jumlah_penderita_dm', 'kategori_dm']].sample(5))

# 18. Hitung Persentase penderita per kota terhadap total tahun tersebut
print("\n--- 18. Menghitung Persentase Kontribusi per Tahun ---")

# Langkah 1: Hitung total per tahun dan tempelkan ke setiap baris data yang sesuai
df['persentase_tahun'] = (df['jumlah_penderita_dm'] / total_tahun_series) * 100

# Tampilkan hasil
print(df[['tahun', 'nama_kabupaten_kota', 'jumlah_penderita_dm', 'persentase_tahun']].head())

# 19. Tabel Ringkas (Tahun, Total DM, Jumlah Kota Tercatat)
print("\n--- 19. Tabel Ringkasan (Summary) ---")
summary_table = df.groupby('tahun').agg({
    'jumlah_penderita_dm': 'sum',      # Total penderita
    'nama_kabupaten_kota': 'count'     # Jumlah kota yg lapor data (count baris)
}).reset_index() # reset_index biar 'tahun' jadi kolom biasa lagi

# Rename kolom biar lebih rapi
summary_table.columns = ['Tahun', 'Total_Penderita', 'Jumlah_Kota_Lapor']
print(summary_table)
```
