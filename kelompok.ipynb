```python
import pandas as pd
# B. PENGOLAHAN DATA (FILTERING & SORTING)
# 1. Load Data
filename = 'dinkes-od_17448_jml_penderita_diabetes_melitus_brdsrkn_kabupatenko_v2_data.csv'
df = pd.read_csv(filename)

# 8. Tampilkan semua baris data untuk tahun 2019 saja
print("\n--- 8. Data Tahun 2019 ---")
data_2019 = df[df['tahun'] == 2019]
print(data_2019)

# 9. Tampilkan kabupaten/kota dengan penderita > 100.000 (tahun berapa pun)
print("\n--- 9. Daerah dengan Kasus Tinggi (>100rb) ---")
# Filter baris di mana kolom jumlah_penderita_dm lebih besar dari 100.000
high_cases = df[df['jumlah_penderita_dm'] > 100000]
print(high_cases)

# 10. Urutkan DataFrame berdasarkan jumlah_penderita_dm (Terbesar ke Terkecil)
print("\n--- 10. Sorting: Penderita Terbanyak (All Time) ---")
sorted_dm = df.sort_values(by='jumlah_penderita_dm', ascending=False)
print(sorted_dm.head()) # Tampilkan 5 teratas biar rapi

# 11. Urutkan berdasarkan Tahun (Asc) lalu Jumlah Penderita (Desc)
print("\n--- 11. Sorting: Tahun & Penderita ---")
# Ini akan mengurutkan tahun dari 2018, 2019... tapi di dalam tahun yg sama,
# kota dengan penderita terbanyak akan muncul duluan.
multi_sort = df.sort_values(by=['tahun', 'jumlah_penderita_dm'], ascending=[True, False])
print(multi_sort)

# 12. Tampilkan 10 kabupaten/kota dengan penderita tertinggi pada tahun 2019
print("\n--- 12. Top 10 Tahun 2019 ---")
# Langkah: Filter 2019 dulu -> Sortir Descending -> Ambil 10 teratas (.head(10))
top_10_2019 = df[df['tahun'] == 2019].sort_values(by='jumlah_penderita_dm', ascending=False).head(10)
print(top_10_2019)

# 13. Tampilkan semua baris data untuk satu kabupaten pilihan (misal: KABUPATEN BOGOR)
print("\n--- 13. History Data: KABUPATEN BOGOR ---")
# Pastikan penulisan nama kota SAMA PERSIS (biasanya huruf kapital semua)
target_kota = 'KABUPATEN BOGOR'
data_bogor = df[df['nama_kabupaten_kota'] == target_kota]
print(data_bogor)
```

```python
import pandas as pd
print("library Import Done cuyy")

f = 'dinkes-od_17448_jml_penderita_diabetes_melitus_brdsrkn_kabupatenko_v2_data.csv'
df = pd.read_csv(f)
print ("Load Data sukses bree ")

# 14.Kelompokkan data berdasarkan tahun dan hitung total jumlah_penderita_dm 
# per tahun di Jawa Barat.
print("\n--- Total Penderita DM per Tahun ---")
total_per_tahun = df.groupby('tahun')['jumlah_penderita_dm'].sum()
print(total_per_tahun)

# 15. Kelompokkan data berdasarkan nama_kabupaten_kota dan hitung 
# rata-rata jumlah_penderita_dm tiap kabupaten/kota (jika terdapat 
# lebih dari satu tahun).
print("\n--- Rata-rata Penderita DM per Kabupaten/Kota ---")
avg_per_kota = df.groupby('nama_kabupaten_kota')['jumlah_penderita_dm'].mean()
print(avg_per_kota)

# 16. Tentukan kabupaten/kota dengan total jumlah_penderita_dm tertinggi 
# dan terendah di seluruh tahun.
print("\n--- Kabupaten/Kota dengan Total DM Tertinggi & Terendah ---")
total_per_kota = df.groupby('nama_kabupaten_kota')['jumlah_penderita_dm'].sum()

print("Tertinggi :", total_per_kota.idxmax(), total_per_kota.max())
print("Terendah  :", total_per_kota.idxmin(), total_per_kota.min())

# 17. Buat kolom baru bernama kategori_dm dengan aturan berbasis jumlah_penderita_dm:
# < 50.000 → "Rendah"
# 50.000–99.999 → "Sedang"
# = 100.000 → "Tinggi"
print("\n--- Membuat Kolom Kategori DM ---")
def tentukan_kategori(jumlah):
    if jumlah >= 100000:
        return "Tinggi"
    elif jumlah >= 50000:
        return "Sedang"
    else:
        return "Rendah"
df['kategori_dm'] = df['jumlah_penderita_dm'].apply(tentukan_kategori)
print(df[['nama_kabupaten_kota', 'jumlah_penderita_dm', 'kategori_dm']].head())

# 18. Untuk setiap tahun, hitung persentase jumlah penderita DM per kabupaten/kota 
# terhadap total jumlah penderita DM di tahun tersebut, lalu simpan dalam kolom baru 
# bernama persentase_tahun.Persentase per tahun
print("\n--- Persentase Kontribusi per Tahun ---")
df['persentase_tahun'] = (
    df['jumlah_penderita_dm'] /
    df.groupby('tahun')['jumlah_penderita_dm'].transform('sum')
) * 100
print(df[['tahun', 'nama_kabupaten_kota',
          'jumlah_penderita_dm', 'persentase_tahun']].head())

# 19. Buat tabel ringkas (misalnya menggunakan groupby) yang menunjukkan:
# tahun, total_jumlah_penderita_dm, dan jumlah_kabupaten_kota 
# yang datanya tercatat di tahun tersebut.
summary_table = df.groupby('tahun').agg({
    'jumlah_penderita_dm': 'sum',
    'nama_kabupaten_kota': 'count'
}).reset_index()
summary_table.columns = ['Tahun', 'Total_Penderita', 'Jml_Kota_Lapor']
print("\n--- Tabel Ringkasan ---")
print(summary_table)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# D. VISUALISASI DATA (MATPLOTLIB)
# 1. Load Data (Persiapan)
filename = 'dinkes-od_17448_jml_penderita_diabetes_melitus_brdsrkn_kabupatenko_v2_data.csv'
df = pd.read_csv(filename)
# --- Re-create Kolom Kategori (Butuh buat Soal No. 23) ---
def tentukan_kategori(jumlah):
    if jumlah >= 100000:
        return 'Tinggi'
    elif jumlah >= 50000:
        return 'Sedang'
    else:
        return 'Rendah'
df['kategori_dm'] = df['jumlah_penderita_dm'].apply(tentukan_kategori)
# ---------------------------------------------------------

# Set Style biar grafik agak cakep 
plt.style.use('ggplot') 

# 20. Grafik Bar: Penderita per Kabupaten/Kota Tahun 2019
print("Membuat Grafik 20...")
data_2019 = df[df['tahun'] == 2019]

plt.figure(figsize=(12, 6))
plt.bar(data_2019['nama_kabupaten_kota'], data_2019['jumlah_penderita_dm'], color='skyblue')
plt.title('Jumlah Penderita DM per Kabupaten/Kota (2019)', fontsize=14)
plt.xlabel('Kabupaten/Kota', fontsize=12)
plt.ylabel('Jumlah Penderita', fontsize=12)
plt.xticks(rotation=90) # Putar label nama kota biar gak tabrakan
plt.tight_layout()
plt.show()

# 21. Grafik Garis: Tren Total Penderita per Tahun
print("Membuat Grafik 21...")
total_per_tahun = df.groupby('tahun')['jumlah_penderita_dm'].sum()

plt.figure(figsize=(10, 6))
plt.plot(total_per_tahun.index, total_per_tahun.values, marker='o', linestyle='-', color='green', linewidth=2)
plt.title('Tren Total Penderita DM di Jawa Barat', fontsize=14)
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Total Penderita', fontsize=12)
plt.grid(True)
plt.xticks(total_per_tahun.index) # Pastikan semua tahun muncul
plt.show()

# 22. Grafik Bar Horizontal: Top 10 Tertinggi Tahun 2019
print("Membuat Grafik 22...")
top_10 = data_2019.sort_values('jumlah_penderita_dm', ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_10['nama_kabupaten_kota'], top_10['jumlah_penderita_dm'], color='salmon')
plt.title('Top 10 Kabupaten/Kota dengan Penderita DM Tertinggi (2019)', fontsize=14)
plt.xlabel('Jumlah Penderita', fontsize=12)
plt.ylabel('Kabupaten/Kota', fontsize=12)
plt.gca().invert_yaxis() # Balik sumbu Y biar ranking 1 ada di atas
plt.tight_layout()
plt.show()

# 23. Pie Chart: Proporsi Kategori DM (2019)
print("Membuat Grafik 23...")
proporsi_kategori = data_2019['kategori_dm'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(proporsi_kategori, labels=proporsi_kategori.index, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('Proporsi Kategori DM (2019)', fontsize=14)
plt.show()

# 24. Grafik Bar: Perbandingan 3 Tahun Terakhir
print("Membuat Grafik 24...")
# Ambil 3 tahun terakhir yang tersedia di data
tahun_tersedia = sorted(df['tahun'].unique())
tiga_tahun_terakhir = tahun_tersedia[-3:] 

# Filter data cuma buat 3 tahun itu
data_3_thn = df[df['tahun'].isin(tiga_tahun_terakhir)]
total_3_thn = data_3_thn.groupby('tahun')['jumlah_penderita_dm'].sum()

plt.figure(figsize=(8, 6))
plt.bar(total_3_thn.index.astype(str), total_3_thn.values, color='cornflowerblue') # astype(str) biar sumbu X gak dianggap angka pecahan
plt.title(f'Total Penderita DM 3 Tahun Terakhir ({tiga_tahun_terakhir})', fontsize=14)
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Total Penderita', fontsize=12)
plt.show()
```
