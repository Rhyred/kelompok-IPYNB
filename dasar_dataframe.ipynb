```python
import pandas as pd
#Import library pandas, baca file data_dm_jabar.csv ke dalam DataFrame bernama df,
#  lalu tampilkan 5 baris pertama.
filename = 'dinkes-od_17448_jml_penderita_diabetes_melitus_brdsrkn_kabupatenko_v2_data.csv'
df = pd.read_csv(filename)

print("--- 1. Lima Baris Pertama ---")
print(df.head())
```

```python
# 2.Tampilkan 5 baris terakhir dari df.
print("\n--- 2. Lima Baris Terakhir ---")
print(df.tail())
```

```python
# 3. Tampilkan informasi struktur DataFrame menggunakan df.info().
print("\n--- 3. Informasi Struktur Data ---")
df.info()
```

```python
# 4. Tampilkan statistik deskriptif untuk kolom numerik 
# (misalnya jumlah_penderita_dm dan tahun) menggunakan df.describe().
print("\n--- 4. Statistik Deskriptif ---")
print(df.describe())
```

```python
# 5. Tampilkan daftar nilai unik dari kolom tahun.
print("\n--- 5. Daftar Tahun Tersedia ---")
print(df['tahun'].unique())
```

```python
# 6. Tampilkan daftar nilai unik dari kolom nama_kabupaten_kota 
# dan hitung berapa banyak kabupaten/kota yang tercatat.
print("\n--- 6. Cek Data Kabupaten/Kota ---")
kota_unik = df['nama_kabupaten_kota'].unique()
jumlah_kota = df['nama_kabupaten_kota'].nunique()

print("Daftar Kota:", kota_unik)
print("Total Kabupaten/Kota:", jumlah_kota)
```
